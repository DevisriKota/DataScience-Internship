 Task 5 - Complete Traffic Accident EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
df1 = pd.read_csv("/content/Task-05 (1).csv")
df2 = pd.read_csv("/content/Task-5 (2).csv")

# Merge on Accident_Index
df = pd.merge(df1, df2, on='Accident_Index')

# Basic stats
print("Dataset Shape:", df.shape)
print(df.describe(include='all'))

# Plot 1: Severity count
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='Severity', palette='Set2')
plt.title('Accident Severity Distribution')
plt.show()

# Plot 2: Gender vs Severity
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Gender', hue='Severity', palette='pastel')
plt.title('Gender-wise Severity')
plt.show()

# Plot 3: Age Group vs Severity
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Age_Grp', hue='Severity', palette='cool')
plt.title('Severity by Age Group')
plt.show()

# Plot 4: Road Conditions
plt.figure(figsize=(10, 5))
sns.countplot(data=df, y='Road_conditions', hue='Severity', palette='magma')
plt.title('Severity by Road Condition')
plt.show()

# Plot 5: Weather Conditions
plt.figure(figsize=(10, 5))
sns.countplot(data=df, y='Weather_conditions', hue='Severity', palette='viridis')
plt.title('Severity by Weather Conditions')
plt.show()

# Plot 6: Time of Day Analysis
df['Hour'] = pd.to_datetime(df['Time'], errors='coerce').dt.hour
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Hour', bins=24, kde=True, color='skyblue')
plt.title('Accidents by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Count')
plt.show()

# Plot 7: Light Conditions
plt.figure(figsize=(10, 5))
sns.countplot(data=df, y='Light_conditions', hue='Severity', palette='rocket')
plt.title('Severity by Lighting Condition')
plt.show()
```
